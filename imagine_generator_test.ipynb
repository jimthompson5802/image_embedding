# Image Generator Test

```python
import tensorflow as tf
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
# location of image files
IMG_DIR = '/Users/jim/Desktop/Deeplearning/transfer_learning/cats_dogs/data/small_train'
```

```python

img_gen = ImageDataGenerator(data_format='channels_last')
```

```python
img_iter = img_gen.flow_from_directory(IMG_DIR,class_mode=None,batch_size=1,
                                      target_size=(128,128),
                                      seed=123)
```
